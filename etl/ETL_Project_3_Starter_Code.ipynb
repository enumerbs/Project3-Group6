```python
# Read the data into a Pandas DataFrame
country_data_df = vaex.from_csv('Metadata_Country_API_NY.GDP.PCAP.CD_DS2_en_csv_v2_3401556.csv', convert=True)
country_data_df.head()
```

```python
# Display the current columns
print("Original Columns:")
print(country_data_df.get_column_names())
```

```python
# Drop unwanted columns using Vaex library
# List the columns you want to drop
columns_to_drop = ['SpecialNotes', 'Unnamed: 5']
```

```python
# Drop the unwanted columns
columns_to_keep = [col for col in country_data_df.get_column_names() if col not in columns_to_drop]
country_data_df = country_data_df[columns_to_keep]
```

```python
#Display new data
country_data_df.head()
```

```python
#Create a Pandas DataFrame

country_data = country_data_df.to_pandas_df()
country_data.head()
```

```python
#Rename Columns - TableName to Country Name
country_data = country_data.rename(columns={'TableName': 'Country Name'})
country_data.head(10)
```

```python
# Load official Country Codes reference data file
country_codes_df = pd.read_csv('CountryCodes_ISO3166.csv')
country_codes_df.head()
```

```python
# Look for 3-letter entries in the dataset that DON'T match official ISO3166 country codes
#
# Return all rows in country_data that do NOT have a matching team in country_codes_df, following the approach in:
# https://www.statology.org/pandas-anti-join/
#
# (1) perform outer join
outer = pd.merge(country_data, country_codes_df, left_on='Country Code', right_on='Alpha-3 code', how='outer', indicator=True)

# (2) perform anti-join
anti_join = outer[(outer._merge=='left_only')].drop('_merge', axis=1)

#view results
anti_join[["Country Name", "Country Code"]]
```

```python
# Extract the (fake) "country codes" so we can use those as a filter
non_country_codes = anti_join["Country Code"]
len(non_country_codes)
```

```python
# Filter out non-country-codes from the original country dataset so we are left with ISO3166 country codes only
new_country_data = country_data[~country_data["Country Code"].isin(non_country_codes)]
new_country_data.head()
```

```python
#Show columns information
new_country_data.info()
```

```python
#Fill Empty Columns

# Create a copy of the DataFrame
new_country_data = new_country_data.copy()
new_country_data.fillna(value=np.nan, inplace=True)
```

```python
#Show columns information after filling empty columns
#nan_values = new_country_data.isnull()
#print(nan_values)

nan_counts = new_country_data.isnull().sum()
print(nan_counts)
```

```python
# Fill the missing value in the 'IncomeGroup' column with 0
new_country_data['IncomeGroup'].fillna(value='Not Specified', inplace=True)
```

```python
# Verify that the missing value has been filled with 0
print(new_country_data['IncomeGroup'].isnull().sum())
```

```python
# Export the DataFrame as a CSV file. 
new_country_data.to_csv("country_data.csv", encoding='utf8', index=False)
```
