```python
import pandas as pd
import janitor
```

```python
# Load the CSV data, specifying that the headers are on row 5 (zero-indexed, so row 4)
uncleaned_df = pd.read_csv("API_SP.POP.GROW_DS2_en_csv_v2_3404396.csv", header=2)

uncleaned_df.head()
```

```python
# Clean column names (headers) to a simpler format using PyJanitor
uncleaned_df = uncleaned_df.clean_names()

uncleaned_df.head()
```

```python
# Count the rows and columns
num_rows, num_columns = uncleaned_df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# Remove duplicate rows
df_no_duplicates = uncleaned_df.drop_duplicates()

# Count the rows and columns after removing duplicates
num_rows, num_columns = df_no_duplicates.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")

df_no_duplicates
```

```python
# Drop the specific 'Indicator Name' and 'Indicator Code' columns
df_no_duplicates = df_no_duplicates.drop(columns=['indicator_name', 'indicator_code'], errors='ignore')
df_no_duplicates
```

```python
# Drop the specific 'unnamed_68' column
df_no_duplicates = df_no_duplicates.drop(columns=['unnamed_68'], errors='ignore')

num_rows, num_columns = df_no_duplicates.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")

df_no_duplicates
```

```python
# Use PyJanitor's `transform_columns` to apply fillna to the '1960' column, replacing NaN with 'Not Available'
df_no_duplicates = df_no_duplicates.transform_columns(['1960'], lambda x: x.fillna('Not Available'))

df_no_duplicates
```

```python
# Load official Country Codes reference data file
country_codes_df = pd.read_csv('CountryCodes_ISO3166.csv')
country_codes_df.head()
```

```python
# (1) perform outer join
outer = pd.merge(df_no_duplicates, country_codes_df, left_on='country_code', right_on='Alpha-3 code', how='outer', indicator=True)

# (2) perform anti-join
anti_join = outer[(outer._merge=='left_only')].drop('_merge', axis=1)

#view results
anti_join[["country_name", "country_code"]]

```

```python
# Extract the (fake) "country codes" so we can use those as a filter
non_country_codes = anti_join["country_code"]
len(non_country_codes)
```

```python
# Filter out non-country-codes from the original population growth dataset so we are left with ISO3166 country codes only
df_no_duplicates = df_no_duplicates[~df_no_duplicates["country_code"].isin(non_country_codes)]
df_no_duplicates
```

```python
# Remove fully blank rows and columns
removed_df = df_no_duplicates.dropna(how='all')  # Drop fully blank rows
removed_df = removed_df.dropna(axis=1, how='all')  # Drop fully blank columns

num_rows, num_columns = removed_df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")

removed_df
```

```python
# Replace missing/blank values with "Not Available"
df_filled = df_no_duplicates.fillna("Not Available")

num_rows, num_columns = df_filled.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")

df_filled.head()
```

```python
# Identify fully blank rows and columns for review
blank_rows = df_no_duplicates[df_no_duplicates.isnull().all(axis=1)]
blank_columns = df_no_duplicates.columns[df_no_duplicates.isnull().all(axis=0)]
```

```python
# Remove fully blank rows and columns
cleaned_df = df_filled.dropna(how='all')  # Drop fully blank rows
cleaned_df = cleaned_df.dropna(axis=1, how='all')  # Drop fully blank columns

num_rows, num_columns = cleaned_df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")

cleaned_df.head()
```

```python
# Save removed rows and columns in another dataframe for review
removed_row_df = pd.DataFrame({"removed_rows": blank_rows.index})  # indices of removed rows
removed_column_df = pd.DataFrame({"removed_columns": blank_columns})   # names of removed columns
```
